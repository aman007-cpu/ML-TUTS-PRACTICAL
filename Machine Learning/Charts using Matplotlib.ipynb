# Styling a Graph

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
```

```python
a1 = [3,6,7,8]
a2 = [12,6,7,3]

b1 = [12,35,19,10]
b2 = [11,22,3,40]
```

```python
plt.plot(a1,b1,'r^--',label='LineOne',linewidth=5)
plt.plot(a2,b2,'c',label='LineTwo',linewidth=5)
plt.title("Basic")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

```python
plt.grid(True,color='k')
```

```python
plt.plot(a1,b1,'g',label='LineOne',linewidth=5)
plt.plot(a2,b2,'c',label='LineTwo',linewidth=5)
plt.title("Basic")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.grid(True,color='k')
plt.legend()
plt.show()
```

# StackPlot


```python
days = [1,2,3,4,5]
sleeping = [7,8,6,12,13]
eating = [2,1,4,3,5]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
```

```python
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)


plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Area Plot')
plt.legend()
plt.show()
```

# Pie Chart

```python
slices = [7,2,2,14]
activities = ['sl','eat','work','play']
cols = ['c','m','r','b']
```

```python
plt.pie(slices , labels = activities,
      colors=cols,startangle=90,shadow=True)
plt.title('Pie Chart')
plt.show()
```
